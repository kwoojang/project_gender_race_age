```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import os
from PIL import Image
```

```python
data = os.listdir('./UTKFace/')
len(data)
```

```python
data = os.listdir('./UTKFace/')
df = pd.DataFrame(data=data, columns=['file_name'])
df
```

```python
drop_list = os.listdir('./0.strange/')
drop_list
```

```python
drop_idx = []
for drop_name in drop_list :
    idx = df[df['file_name'] == drop_name].index[0]
    drop_idx.append(idx)
drop_idx
```

```python
print(len(drop_list))
len(drop_list) == len(drop_idx)
```

```python
df
```

```python
df.drop(index=drop_idx, inplace=True)
df
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
df['age'] = df['file_name'].apply(lambda x : int(x.split('_')[0]))
df['gender'] = df['file_name'].apply(lambda x : int(x.split('_')[1]))
```

```python
df['gender'].unique()
```

```python
df['race'] = df['file_name'].apply(lambda x : x.split('_')[2])
```

```python
df['race'].unique()
```

```python
df = df[(df['race'] == '0') | (df['race'] == '1') | (df['race'] == '2')].reset_index(drop=True)
```

```python
df
```

```python
df['race'] = df['race'].apply(lambda x : int(x))
df
```

```python
df.info()
```

```python
plt.pie(list(df['gender'].value_counts()), labels=['Men', 'Women'], 
       autopct='%.1f%%', explode=[0.01, 0.01], counterclock=False, shadow=True)
plt.title('Gender')
plt.show()
```

```python
df['age_cat'] = df['age'].apply(lambda x : x // 10)
df
```

```python
mapper = {0 : '<20',
          1 : '<20',
          2 : '20',
          3 : '30',
          4 : '40',
          5 : '50',
          6 : '60',
          7 : '70',
          8 : '80+',
          9 : '80+',
          10 : '80+',
          11 : '80+'}
```

```python
df['age_cat'] = df['age_cat'].map(mapper)
df
```

```python
plt.pie(list(df['age_cat'].value_counts()), labels=list(df['age_cat'].value_counts().index), 
       autopct='%.1f%%', explode=([0.01] * df['age_cat'].nunique()), counterclock=False, shadow=True)
plt.title('Age')
plt.show()
```

```python
plt.pie(list(df['race'].value_counts()), labels=['White', 'Black', 'Asian'], 
       autopct='%.1f%%', explode=([0.01] * df['race'].nunique()), counterclock=False, shadow=True)
plt.title('Race')
plt.show()
```

```python
df[df['race'] == 2]
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
rus = RandomUnderSampler()

df_under, data_label = rus.fit_resample(df, df['race'])
```

```python
df_under = pd.DataFrame(data=df_under, columns=df.columns).reset_index(drop=True)
df_under
```

```python
plt.pie(list(df_under['race'].value_counts()), labels=['White', 'Black', 'Asian'], 
       autopct='%.1f%%', explode=([0.01] * df_under['race'].nunique()), counterclock=False, shadow=True)
plt.title('Race')
plt.show()
```

```python
plt.pie(list(df_under['gender'].value_counts()), labels=['Men', 'Women'], 
       autopct='%.1f%%', explode=[0.01, 0.01], counterclock=False, shadow=True)
plt.title('Gender')
plt.show()
```

```python
plt.pie(list(df_under['age_cat'].value_counts()), labels=list(df_under['age_cat'].value_counts().index), 
       autopct='%.1f%%', explode=([0.01] * df_under['age_cat'].nunique()), counterclock=False, shadow=True)
plt.title('Age')
plt.show()
```
